```sql
/****** Object:  StoredProcedure [dbo].[ObtenerEstados]    Script Date: 29/04/2022 1:55:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [dbo].[ObtenerEstados]

AS
BEGIN
	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	/****** Script for SelectTopNRows command from SSMS  ******/
	SELECT *, DATEDIFF(DAY,estado.FechaPrestamo, estado.FechaDevolucion) as dias
	  FROM [Libreria].[dbo].[EstadosPrestamos] estado
                                            join Estudiantes est on estado.IDEstudiante = est.Id 
                                            join libros lib on estado.IDLibro = lib.Id
END
GO
```
